# BM25 Retriever Smoke Test

This notebook builds a small BM25 index with 10 entries and retrieves the top 2 matches for a sample query.

```python
import tempfile
from pathlib import Path

import nltk
import pandas as pd

from llm_bayesian_reasoning.retrievers.retrievers import BM25Retriever

nltk.download("punkt", quiet=True)
nltk.download("punkt_tab", quiet=True)
```

```python
entities = [
    "Cats often sleep on warm windowsills during the afternoon.",
    "Dogs enjoy running through parks and chasing tennis balls.",
    "A domestic cat likes warm blankets and quiet places to nap.",
    "Basketball players practice free throws after team training.",
    "The Eiffel Tower is one of the most famous landmarks in Paris.",
    "Fresh bread is baked every morning in the neighborhood bakery.",
    "Machine learning models can classify text into categories.",
    "Mountaineers prepare carefully before climbing snowy peaks.",
    "A cat curled up beside the fireplace for an evening nap.",
    "Students reviewed probability notes before the exam.",
]

titles = [f"doc_{index}" for index in range(len(entities))]
query = "Where does a cat like to nap in a warm place?"

temp_dir = tempfile.TemporaryDirectory(prefix="bm25_test_index_")
index_path = Path(temp_dir.name)

retriever = BM25Retriever()
retriever.build_index(entities=entities, titles=titles, index_path=index_path)

print(f"Index written to: {index_path}")
print(f"Indexed {len(entities)} entries")
```

```python
top_k = 2
retrieval = retriever.retrieve(query, top_k=top_k)
raw_results = retrieval.results
documents = retrieval.documents

print(f"Query: {query}")
print("\nRaw top-2 results (index, score):")
print(raw_results)

pd.DataFrame([doc.model_dump() for doc in documents])
```
